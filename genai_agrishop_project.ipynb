```python
from transformers import pipeline
```

```python
agri_bot = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)
```

```python
question = "Suggest fertilizer for wheat crop"
response = agri_bot(question)

print(response[0]['generated_text'])
```

```python
def agri_shop_assistant(query):
    result = agri_bot(query)
    return result[0]['generated_text']
```

```python
queries = [
    "Which pesticide is best for cotton pests?",
    "How to increase rice yield?",
    "Best seeds for tomato farming",
    "Organic fertilizer for vegetables"
]

for q in queries:
    print("Question:", q)
    print("Answer:", agri_shop_assistant(q))
    print("-" * 50)
```

```python
while True:
    user_question = input("Ask Agri Question (type 'exit' to stop): ")
    if user_question.lower() == "exit":
        break
    print("AgriBot:", agri_shop_assistant(user_question))
```
